# Comparaison inter-domaines

On exécute le **même** pipeline sur les trois domaines (industrie, santé,
aérien) via `scripts/run_cross_domain.py`, puis on affiche le tableau
comparatif. Données synthétiques par défaut : valeurs illustratives.


```python
import subprocess
import sys
from pathlib import Path

ROOT = Path.cwd().parent
sys.path.insert(0, str(ROOT / "src"))
```

```python
# Lance le pipeline sur les 3 domaines (génère des données synthétiques).
subprocess.run(
    [
        sys.executable,
        str(ROOT / "scripts" / "run_cross_domain.py"),
        "--epochs",
        "3",
        "--image-size",
        "64",
    ],
    cwd=ROOT,
    check=True,
)
```

```python
import json
import pandas as pd

data = json.loads((ROOT / "reports" / "metrics" / "cross_domain.json").read_text())
df = pd.DataFrame(data)[
    ["domain", "split_rule", "has_pixel_masks", "n_test", "auroc", "f1", "fn", "fp"]
]
df
```

## Lecture

- **Ce qui est commun** : preprocessing, autoencodeur, entraînement,
  évaluation — code identique pour les trois domaines.
- **Ce qui change** : la règle de split (par patient en santé, par zone en
  aérien) et une fine classe dataset.
- La comparaison des AUROC entre domaines doit rester **prudente** :
  datasets, résolutions et types d'anomalie diffèrent. Sur données
  synthétiques, les chiffres n'ont qu'une valeur illustrative.
